3. Ubah ke matriks 4x5
4. Tampilkan shape dan transpose

## Penjelasan Bagian NumPy

Pada bagian ini membuat 20 angka acak menggunakan fungsi np.random.rand(20). Angka yang dihasilkan awalnya berada pada rentang 0 sampai 1, kemudian dikalikan 100 agar nilainya menjadi antara 0 hingga 100.

Data tersebut kemudian ditampilkan dengan pembulatan dua angka di belakang koma menggunakan round(2) supaya hasilnya lebih rapi dan mudah dibaca.

Setelah itu  menghitung beberapa ukuran statistik dasar seperti rata-rata (mean), nilai tengah (median), standar deviasi (std), nilai minimum, dan nilai maksimum. Perhitungan ini dilakukan untuk mengetahui gambaran umum dari sebaran data yang telah dibuat.

Selanjutnya, array yang awalnya satu dimensi diubah menjadi bentuk matriks berukuran 4 baris dan 5 kolom menggunakan fungsi reshape(4,5). Perubahan bentuk ini bertujuan untuk memahami bagaimana data dapat disusun dalam bentuk tabel atau matriks.

Terakhir,  menampilkan ukuran matriks menggunakan shape untuk mengetahui dimensinya, serta menggunakan transpose (.T) untuk menukar posisi baris dan kolom. Langkah ini penting karena dalam machine learning, data sering diolah dalam bentuk matriks dengan berbagai ukuran.

```python
arr = np.random.rand(20) * 100

print("Array (20 angka):")
print(arr.round(2))

mean_val = arr.mean()
median_val = np.median(arr)
std_val = arr.std()
min_val = arr.min()
max_val = arr.max()

print("\nStatistik:")
print("Mean  :", round(mean_val, 2))
print("Median:", round(median_val, 2))
print("Std   :", round(std_val, 2))
print("Min   :", round(min_val, 2))
print("Max   :", round(max_val, 2))

mat = arr.reshape(4, 5)

print("\nMatriks 4x5:")
print(mat.round(2))
print("\nShape matriks:", mat.shape)
print("\nTranspose:")
print(mat.T.round(2))
```

2. Tambahkan kolom Status: Lulus jika nilai ≥ 75, selain itu Tidak Lulus
3. Tampilkan:
   - 3 nilai tertinggi
   - Rata-rata nilai kelas
   - Jumlah mahasiswa lulus

## Penjelasan Bagian Pandas ( Data Frame Mahasiswa)
Pada bagian ini  membuat daftar nama mahasiswa secara otomatis menggunakan perulangan dari 1 sampai 10. Jadi tidak perlu menulis satu per satu, karena sudah dibuat dengan format Mahasiswa_1, Mahasiswa_2, dan seterusnya.

Nilai mahasiswa dibuat secara acak menggunakan np.random.randint(50, 101, size=10). Artinya, sistem akan menghasilkan 10 angka dengan rentang nilai antara 50 sampai 100.

Setelah itu, nama dan nilai digabungkan ke dalam sebuah DataFrame menggunakan pd.DataFrame(). DataFrame ini berfungsi seperti tabel, yang memiliki kolom “Nama” dan “Nilai”.

Selanjutnya menambahkan kolom baru bernama “Status”. Kolom ini dibuat menggunakan np.where() untuk memberikan keterangan:

Jika nilai lebih dari atau sama dengan 75 maka statusnya “Lulus”.

Jika kurang dari 75 maka statusnya “Tidak Lulus”.

Terakhir, DataFrame ditampilkan untuk melihat hasil tabel mahasiswa beserta nilai dan status kelulusannya.

```python
nama = [f"Mahasiswa_{i+1}" for i in range(10)]
nilai = np.random.randint(50, 101, size=10)  # 50–100

df_mhs = pd.DataFrame({
    "Nama": nama,
    "Nilai": nilai
})

df_mhs["Status"] = np.where(df_mhs["Nilai"] >= 75, "Lulus", "Tidak Lulus")

print("DataFrame Mahasiswa:")
df_mhs
```

## Penjelasan Bagian Pandas ( Mencari Nilai Tertinggi Mahasiswa)
df_mhs.nlargest(3, "Nilai")
Digunakan untuk mengambil tiga mahasiswa dengan nilai tertinggi. Fungsi nlargest() secara otomatis mengurutkan data dari nilai terbesar lalu mengambil jumlah data yang diminta.

display(top3)
Digunakan untuk menampilkan hasil tiga nilai tertinggi dalam bentuk tabel agar lebih rapi dan mudah dibaca.

df_mhs["Nilai"].mean()
Berfungsi untuk menghitung rata-rata nilai seluruh mahasiswa dalam satu kelas. Perhitungan ini membantu melihat gambaran umum performa kelas.

round(rata2, 2)
Digunakan untuk membulatkan hasil rata-rata menjadi dua angka di belakang koma supaya tampilannya lebih jelas.

(df_mhs["Status"] == "Lulus").sum()
Digunakan untuk menghitung jumlah mahasiswa yang lulus. Cara kerjanya adalah memeriksa setiap baris pada kolom “Status”, lalu menjumlahkan yang bernilai “Lulus”.

int(jml_lulus)
Digunakan untuk memastikan hasil yang ditampilkan berupa angka bulat tanpa desimal.

```python
top3 = df_mhs.nlargest(3, "Nilai")
print("Top 3 Nilai:")
display(top3)

rata2 = df_mhs["Nilai"].mean()
print("\nRata-rata nilai kelas:", round(rata2, 2))

jml_lulus = (df_mhs["Status"] == "Lulus").sum()
print("Jumlah mahasiswa lulus:", int(jml_lulus))
```
